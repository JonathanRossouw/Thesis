```python
# 3.00E-13 - critical!

 

df=pd.read_csv('rho_liquidity/data.csv')

iden = 'AV'

tr=df[df.current_step==8]

M=tr[[iden,'illiquidity', 'shock']]

M = M[M.illiquidity<1.00E-12]

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
import matplotlib.ticker as mtick

 
 

fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot using `.trisurf()`:

surf=ax.plot_trisurf(  M[iden]*100,  M['shock']*100, M['illiquidity'],cmap=cm.coolwarm, linewidth=0.05)
ax.view_init(40, 100) #40/200
ax.set_xlabel('\n Spillover losses in percent of pre-shock equity', fontsize=8)
ax.set_ylabel('\n shock size in %', fontsize=8)
ax.set_zlabel('\n illiquidity', fontsize=8)

ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
# ax.yaxis.set_major_formatter(mtick.PercentFormatter())

# plt.title('Aggregate Vulnerability \n')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
